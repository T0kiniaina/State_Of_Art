#State of Art Google scholar

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
df = pd.read_csv(r'/content/drive/My Drive/Data_analytics/Projects/20241008_state_of_art/Google Scholar.csv')
df
```

```python
# Convert the type of the values in the "Cite" column to int.
df["Cite"]=df["Cite"].astype(int)

#pivot_table
year = 2020  # Filter for the year 2020 in the table.

filtered_df = df[pd.to_datetime(df['Year'], errors='coerce').dt.year == 2020]
df0 = filtered_df.copy()

# Create a pivot table to facilitate the visualization
df0 = pd.pivot_table(
    df,
    index=["Author1"],
    values=["Cite"],
    aggfunc="mean"
)
df0 = df0.sort_values(by='Cite', ascending=False)
df0
```

```python
# Remove the outlier values and the null cites.
df0 = df0.iloc[2:]
df0 = df0[df0['Cite'] != 0]
df0 = df0.head(50)
df0
```

```python
import matplotlib.pyplot as plt

# Assuming 'df0' is your DataFrame

# Create a bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.bar(df0.index, df0['Cite'], color="#045b0b")
plt.gcf().set_facecolor('#b8e2d5')
plt.gca().set_facecolor('#dae3e0')
plt.xlabel("Author1")
plt.ylabel("Mean Cite")
plt.title("Mean Citations by Author1 (Excluding Outliers and Zeros)")
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```
